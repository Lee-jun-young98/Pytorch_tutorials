# 덧셈4

   - (참고) in-place 방식
      - in-place 방식으로 텐서의 값을 변경하는 연산 뒤에는 "_"가 붙음
      - x.copy_(y), x.t_()

```python
print(x)
print(y)
y.add_(x) # y += x
print(y)
```

# 그 외의 연산
  - torch.sub : 뺄셈
  - torch.mul : 곱셈
  - torch.div : 나누셈
  - torch.mm : 내적(dot product)

```python
x = torch.Tensor([[1,3],
                  [5,7]])
y = torch.Tensor([[2,4],
                  [6,8]])
print(x-y)
print(torch.sub(x,y))
print(x.sub(y))
```

```python
x = torch.Tensor([[1,3],
                  [5,7]])
y = torch.Tensor([[2,4],
                  [6,8]])
print(x*y)
print(torch.mul(x,y))
print(x.mul(y))
```

```python
x = torch.Tensor([[1,3],
                  [5,7]])
y = torch.Tensor([[2,4],
                  [6,8]])
print(x/y)
print(torch.div(x,y))
print(x.div(y))
```

```python
x = torch.Tensor([[1,3],
                  [5,7]])
y = torch.Tensor([[2,4],
                  [6,8]])

print(torch.mm(x,y))
```

# 차원을 축소(제거)

```python
tensor = torch.rand(1, 3, 3)
tensor.shape

print(tensor)
```

```python
t = tensor.squeeze()

print(t)
print(t.shape)
```

```python
t = tensor.unsqueeze(dim = 0) # 차원을 증가 

print(t)
print(t.shape)
```
