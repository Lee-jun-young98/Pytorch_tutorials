# 더 수학적 텐서
https://pytorch.org/docs/stable/torch.html#math-operations

```python
a = torch.rand(2, 4) * 2 - 1
print('Common functions:')
print(torch.abs(a)) # 절대값
print(torch.ceil(a)) # 각 요소보다 크거나 같은 최소 정수인 입력 요소의 ceil이 포함된 새 텐서를 반환합니다.
print(torch.floor(a)) # 입력 요소의 바닥이 있는 새 텐서를 반환합니다. 각 요소보다 작거나 같은 최대 정수입니다.
print(torch.clamp(a, -0.5, 0.5)) # 입력된 모든 요소를 [min, max] 범위로 클램프합니다.
```

```python
# 삼각함수와 역함수
angles = torch.tensor([0, math.pi / 4, math.pi / 2, 3 * math.pi / 4])
sines = torch.sin(angles)
inverses = torch.asin(sines)
print('\nSine and arcsine:')
print(angles) # 각 입력 요소가 라디안 단위의 각도에서 각도로 변환된 새 텐서를 반환합니다.
print(sines) # 입력 요소의 사인 값을 사용하여 새 텐서를 반환합니다.
print(inverses) # 입력 요소의 asin이 포함된 새 텐서를 반환합니다.
```

```python
# bitwise operations 입력 및 기타 비트 단위 XOR를 계산합니다.
print('\nBitwise XOR:')
b = torch.tensor([1, 5, 11])
c = torch.tensor([2, 7, 10])
print(torch.bitwise_xor(b, c))
```

```python
# comparisons
print('\nBroadcasted, element-wise equality comparison: ')
d = torch.tensor([[1., 2.], [3., 4.]])
e = torch.ones(1, 2) 
print(torch.eq(d, e)) # 행렬 비교 연산 후 boolean 타입 반환
```

```python
# reductions:
print('\nReduction ops:')
print(torch.max(d)) # 가장 큰 싱글 텐서 리턴
print(torch.max(d).item()) # 큰 싱글 텐서 값  추출
print(torch.mean(d)) # 평균
print(torch.std(d)) # 표준편차
print(torch.prod(d)) # 모든 숫자 곱
print(torch.unique(torch.tensor([1, 2, 1, 2, 1, 2]))) # 유니코한 요소 리턴
```

```python
# 벡터와 선형 연산
v1 = torch.tensor([1., 0., 0.]) # x 축 벡터
v2 = torch.tensor([0., 1., 0.])
m1 = torch.rand(2, 2)
m2 = torch.tensor([[3., 0.], [0., 3.]]) # 3차원 매트릭스

print('\nVectors & Matrices:')
print(torch.cross(v2, v1)) # 벡터의 교차곱을 반환
print(m1)
m3 = torch.matmul(m1, m2)
print(m3)
print(torch.svd(m3)) # 특이값 분해
```

```python
# 텐서 교체
a = torch.tensor([0, math.pi / 4, math.pi / 2, 3 * math.pi / 4])
print('a:')
print(a)
print(torch.sin(a))
print(a)

b = torch.tensor([0, math.pi / 4, math.pi / 2, 3 * math.pi / 4])
print('\nb: ')
print(b)
print(torch.sin_(b))
print(b)
```

```python
a = torch.ones(2, 2)
b = torch.rand(2, 2)

print('Before: ')
print(a)
print(b)
print('\nAfter adding:')
print(a.add_(b))
print(a)
print(b)
print('\nAfter multiplying')
print(b.mul_(b))
print(b)
```

```python
a = torch.rand(2, 2)
b = torch.rand(2, 2)
c = torch.zeros(2, 2)
old_id = id(c)

print(c)
d = torch.matmul(a, b, out=c)
print(c)

assert c is d # c와 d가 동일한 객체 인지 확인
assert id(c), old_id # 새로운 c가 이전 객체와 동일한지 확인

torch.rand(2, 2, out=c)
print(c) # c가 다시 바뀜
assert id(c), old_id # 여전히 똑같은 객체
```

# 텐서 복제

```python
a = torch.ones(2, 2)
b = a

a[0][1] = 561
print(b)
```

```python
a = torch.ones(2, 2)
b = a.clone()

assert b is not a # 다른 메모리에 복제함
print(torch.eq(a, b)) # 같은 객체

a[0][1] = 561
print(b)
```

소스텐서가 autograd를 활성화하면 클론도 활성화 된다

```python
a = torch.rand(2, 2, requires_grad=True)
print(a)

b = a.clone()
print(b)

c = a.detach().clone()
print(c)

print(a)
```

# Numpy Bridge
- numpy array와 pytorch 변환이 쉽다

```python
import numpy as np

numpy_array = np.ones((2, 3))
print(numpy_array)

pytorch_tensor = torch.from_numpy(numpy_array)
print(pytorch_tensor)
```

```python
pytorch_rand = torch.rand(2, 3)
print(pytorch_rand)

numpy_rand = pytorch_rand.numpy()
print(numpy_rand)
```

```python
numpy_array[1, 1] = 23
print(pytorch_tensor)

pytorch_rand[1, 1] = 17
print(numpy_rand)
```
